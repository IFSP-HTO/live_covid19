## O modelo S.I.R.

Kermack e A. G. McKendrick (1927) criaram um modelo em que consideravam uma população fixa com apenas três conjuntos ( compartimentos):

$ S$,  $I$ e $R$

Em cada "instante" $t$

$S(t)$ é o número de indivíduos ainda não infetados com a doença, ou aqueles sensíveis, suscetíveis à doença.

$I(t)$ é o número de indivíduos que foram infetados com a doença e são capazes de propagar a doença para aqueles na categoria suscetível.

$R(t)$ é o número de indivíduos que foram infectados recuperados da doença e adquirem imunidade.

## um modelo com dois parâmetros

$$ \frac{d S}{dt} = -\beta \, I \, S $$

$$ \frac{d I}{dt} = \beta \, S \, I \, - \gamma \, I $$

$$ \frac{d R}{dt} = \gamma \, I $$

$\beta$ é a taxa de contato ou de infeção da doença. $\gamma$ é a taxa de recuperação. 

$1/\gamma$ é período de infecção/contágio/recuperação. $R_0=\beta/\gamma$ é a taxa de reprodutividade da epidemia. 

## Computação Científica - Python

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# Population de referência, N. Um milhão
N = 1000000 #########
# I(0) e R(0):
I0, R0 = 0.08, 0.0 
# S(0).
S0 = N - I0 - R0
# beta e gamma, (tempo em dias).
beta, gamma = 0.32, 0.16 #################
# dias
tf=200;
t = np.linspace(0, tf, tf)

# The SIR model differential equations.
def deriv(y, t, N, beta, gamma):
    S, I, R = y
    dSdt = -beta * S * I / N
    dIdt = beta * S * I / N - gamma * I
    dRdt = gamma * I
    return dSdt, dIdt, dRdt

# Initial conditions vector
y0 = S0, I0, R0
Rnot=beta/gamma; # 
# Integrate the SIR equations over the time grid, t.
ret = odeint(deriv, y0, t, args=(N, beta, gamma))
S, I, R = ret.T

# Plot the data on three separate curves for S(t), I(t) and R(t)
fig = plt.figure(facecolor='w')
ax = fig.add_subplot(111,  axisbelow=True)
ax.plot(t, S, 'b', alpha=0.5, lw=2, label='Susceptível')
ax.plot(t, I, 'r', alpha=0.5, lw=2, label='Infectado')
ax.plot(t, R, 'g', alpha=0.5, lw=2, label='Recuperado')
ax.set_xlabel('Tempo /dias')
ax.set_ylabel('Número por milhão')
ax.set_ylim(0,N)
ax.yaxis.set_tick_params(length=0)
ax.xaxis.set_tick_params(length=0)
ax.grid(b=True, which='major', c='w', lw=2, ls='-')
legend = ax.legend()
legend.get_frame().set_alpha(0.5)
for spine in ('top', 'right', 'bottom', 'left'):
    ax.spines[spine].set_visible(False)
plt.show()
```
